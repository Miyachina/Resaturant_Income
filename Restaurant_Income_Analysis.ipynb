```python
import numpy as np
```

```python
import pandas as pd
from numpy.random import randn
from scipy import stats
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
fast = sns.load_dataset('tips')
```

```python
fast.head()
```

```python
sns.lmplot('total_bill','tip', fast)
```

```python
sns.lmplot('total_bill','tip',fast, scatter_kws = {'marker':'c','color':'indianred'}, line_kws = {'linewidth':1,'color':'blue'})
```

```python
sns.lmplot('total_bill','tip',fast, order = 4, scatter_kws = {'marker':'c','color':'indianred'}, line_kws = {'linewidth':1,'color':'blue'})
```

```python
# let's examine he scatter plot without regression line
sns.lmplot('total_bill','tip',fast, fit_reg = False)
```

```python
# create a new column for tip percentage
fast['tip_pect'] = 100*(fast['tip']/fast['total_bill'])
```

```python
fast.head()
```

```python
sns.lmplot('size','tip_pect',fast)
```

```python
sns.lmplot('size','tip_pect',fast, x_jitter=.1)
```

```python
# estimate the tendency of each bin, each bin is the size of the party in this case
sns.lmplot('size','tip_pect',fast, x_estimator=np.mean)
```

```python
sns.lmplot('total_bill','tip_pect',fast,hue='sex',markers=['x','o'])
```

```python
#does a day inflences the amount of tip paid?
sns.lmplot('total_bill','tip_pect',fast,hue='day')
```

```python
sns.lmplot('total_bill','tip_pect',fast,lowess=True, line_kws={'color':'black'})
#local regression in sns
```

```python
fig, (axis1,axis2) = plt.subplots(1,2,sharey=True)

sns.regplot('total_bill','tip_pect',fast, ax=axis1)
sns.violinplot(fast['size'],fast['tip_pect'],color='Red', ax=axis2)
```
